## Workhours

```python
import numpy as np
import pandas as pd
import datetime
from datetime import datetime
import networkx as nx
```

```python
# load the dataset
df=pd.read_csv('../Data/pageneral.csv')
```

```python

#Strip white spaces
df=df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df['Originator']=df['Originator'].str.extract('(\d+)').astype(int)
```

```python
df['Recipient']=df['Recipient'].str.extract('(\d+)').astype(int)
```

```python
# load the dataset
ncs=pd.read_csv('../Data/ncs.csv')
```

```python
#The Merger
df=pd.merge(df, ncs, on='Originator')
```

```python
# load the Workhours dataset
wh=pd.read_csv('../Data/WorkHours.csv')
```

```python
#Strip white spaces
wh=wh.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
wh['Originator']=wh['Originator'].str.extract('(\d+)').astype(int)
```

```python
#The Merger
wh1=pd.merge(df, wh, on='Originator')
```
