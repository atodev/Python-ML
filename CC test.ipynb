```python
#import libraries
import numpy as np
import pandas as pd
import datetime as dt
```

```python
#load the pre data
dfpr=pd.read_csv('Acc-Covid-wth-nw-pre.csv')
```

```python
#load the post data
dfpo=pd.read_csv('Acc-Covid-wth-nw-post.csv')
```

```python
dfpr[["datetime"]] = dfpr[["datetime"]].apply(pd.to_datetime)
dfpo[["datetime"]] = dfpo[["datetime"]].apply(pd.to_datetime)
#collect weekdays from datetime
dfpr['day_of_week'] = dfpr['datetime'].dt.day_name()
dfpo['day_of_week'] = dfpo['datetime'].dt.day_name()
#collect hours from datetime
dfpr['hour'] = dfpr['datetime'].dt.strftime('%H').astype(int)
dfpo['hour'] = dfpo['datetime'].dt.strftime('%H').astype(int)
```

```python
PreCC=dfpr.groupby ('address')['out'].count().sort_values(ascending=False)
PreCC
```

```python
PostCC=dfpo.groupby ('address')['out'].count().sort_values(ascending=False)
PostCC
```

```python
PrePo=pd.merge(PreCC, PostCC, on='address')
PrePo
```

```python

## Remove duplicate datetime  from address to eliminate cc's
dfpr = dfpr.reset_index().drop_duplicates(subset=['address','datetime'],keep='first').set_index('index')
dfpo = dfpo.reset_index().drop_duplicates(subset=['address','datetime'],keep='first').set_index('index')
```

```python
PreNCC=dfpr.groupby ('address')['out'].count().sort_values(ascending=False)
PreNCC
```

```python
PostNCC=dfpo.groupby ('address')['out'].count().sort_values(ascending=False)
PostNCC
```

```python
PrePos3=pd.merge(PrePo, PreNCC, on='address')
PrePos3
```

```python
PrePos4=pd.merge(PrePos3, PostNCC, on='address')
PrePos4
```

```python
PrePos4.to_csv(r'PrePos5.csv')
```
